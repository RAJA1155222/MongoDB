Q1. What is MongoDB? Explain non-relational databases in short.

In which scenarios is MongoDB preferred over SQL databases?**

MongoDB is a NoSQL (non-relational), document-oriented database that stores data in flexible, JSON-like documents.

👉 Non-relational databases (in short)

Non-relational (NoSQL) databases store data without tables and fixed schemas.
They use structures such as documents, key-value pairs, or graphs.

👉 When MongoDB is preferred over SQL databases

MongoDB is preferred when:

the data structure changes frequently

the application needs high scalability

the data is semi-structured or unstructured

fast development is required without designing complex schemas

the system must handle large volumes of data

Q4. Insert one record, insert many records and use find() and find_one()

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["school"]
collection = db["students"]

# insert one record
collection.insert_one({
    "roll": 1,
    "name": "Amit",
    "age": 20
})

# insert many records
collection.insert_many([
    {"roll": 2, "name": "Ravi", "age": 21},
    {"roll": 3, "name": "Neha", "age": 22}
])

# find one record
print("One record:")
print(collection.find_one())

# find all records
print("All records:")
for doc in collection.find():
    print(doc)
```

Q5. Explain how the find() method is used to query MongoDB.

Write a simple code.**

The find() method is used to retrieve documents that match a given condition (filter).

Syntax:

collection.find({filter})

Example – fetch students whose age is 21:

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["school"]
collection = db["students"]

for doc in collection.find({"age": 21}):
    print(doc)
```

Q6. Explain the sort() method. Give an example.

The sort() method is used to arrange the result documents in ascending or descending order.

Syntax:

collection.find().sort(field, order)

1 → ascending

-1 → descending

Example – sort students by age in ascending order:

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["school"]
collection = db["students"]

for doc in collection.find().sort("age", 1):
    print(doc)
```

Used to delete all documents that match a condition.

Example:
